## Finding the outliers

```python
numerical = df.select_dtypes(exclude = 'object')
```

```python
for i in numerical.columns:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    IQR = q3 - q1
    lower_limit = q1 - (1.5 * IQR)
    upper_limit = q3 + (1.5 * IQR)
    print(f"{i} lower_ limit {lower_limit}")
    print(f"{i} upper_limit {upper_limit}")
    print()
```

```python
for i in numerical.columns:
    q1 = numerical[i].quantile(0.25)
    q3 = numerical[i].quantile(0.75)
    IQR = q3 - q1
    lower_limit = q1 - (1.5 * IQR)
    upper_limit = q3 + (1.5 * IQR)
    lower = numerical[numerical[i]<lower_limit] 
    upper = numerical[numerical[i]>upper_limit] 
    print(f"{i} lower_ limit {lower}")
    print(f"{i} upper_limit {upper}")
    print()
```

## Treatment for outliers


```python
for i in numerical.columns:
    q1 = df[i].quantile(0.25)
    q3 = df[i].quantile(0.75)
    IQR = q3 - q1
    lower_limit = q1 - (1.5 * IQR)
    upper_limit = q3 + (1.5 * IQR)
    condition = (df[i]<lower_limit) | (df[i]>upper_limit)
    df[i] = np.where(condition,df[i].mode()[0],df[i])
    
```

```python
df['price'].plot(kind = 'box')  ## after treatment there is no outlier in data 
```
