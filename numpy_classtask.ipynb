```python
import numpy as np
```

```python
data = np.genfromtxt(r"C:\Users\mm732\Downloads\Iris_clean_-_Copy.csv",dtype = "float16",delimiter = ",")
```

```python
data
```

```python
sepal_length = data[1:,0:1]
```

```python
sepal_length
```

```python
np.where(sepal_length>10) # outliers
```

```python
sepal_length = np.where(sepal_length>10,np.nanmedian(sepal_length),sepal_length)  #  outliers conversion
```

```python
sepal_length
```

```python
sepal_length = np.where(np.isnan(sepal_length),np.nanmean(sepal_length),sepal_length)  # conversion of missing values
```

```python
sepal_length
```

```python
sepal_width = data[1:,1:2]
```

```python
sepal_width
```

```python
np.where(sepal_width>5) # outliers
```

```python
sepal_width = np.where(sepal_width>5,np.nanmedian(sepal_width),sepal_width) # conversion of outliers  
```

```python
sepal_width
```

```python
sepal_width = np.where(np.isnan(sepal_width),np.nanmean(sepal_width),sepal_width)  # conversion of missing values 
```

```python
sepal_width
```

```python
petal_length = data[1:,2:3]
```
